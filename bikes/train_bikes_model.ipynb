```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
import pickle as pkl
import seaborn as sns
sns.set()
np.random.seed(0)
```

```python
df = pd.read_csv('data/day.csv', index_col=0)
df.head()
```

```python
df.describe().T
```

```python
df2 = df.loc[:,('season',
                'yr',
                'mnth',
                'holiday',
                'weekday',
                'workingday',
                'weathersit',
                'temp',
                'atemp',
                'hum',
                'windspeed',
                'cnt')]

df2.rename(columns={
                    'yr': 'year',
                    'mnth': 'month',
                    'weekday': 'week_day',
                    'workingday': 'working_day',
                    'weathersit': 'weather',
                    'temp': 'temperature',
                    'atemp': 'feeling_temperature',
                    'hum': 'humidity',
                    'windspeed': 'wind_speed',
                    'cnt': 'y'
                  }, inplace=True)
df2.head()
```

```python
X = df2
y = X.pop('y')
X.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

lr_pipeline = Pipeline([
    ('lr', LinearRegression())
])

lr_pipeline.fit(X_train.values, y_train.values)

print(f"Score: {lr_pipeline.score(X_test.values, y_test.values)}")
```

```python
X_train['y'] = y_train
X_train.to_csv("bikes_train.csv")
X_test['y'] = y_test
X_test.to_csv("bikes_test.csv")
```

```python
with open("bikes_model_linear_regression.pkl", "wb") as f:
    pkl.dump(lr_pipeline, f)
```
